```python
import pandas as pd
```

```python
#Read in each of the csv files for merging

w_df=pd.read_csv('cities_weather_data_cleaned_us_only.csv')
b_df=pd.read_csv('BEA_data_for_cross_reference.csv')
i_df=pd.read_csv('inbound_data_for_cross_reference.csv')
q_df=pd.read_csv('qol_cleaned_us_only.csv')
```

```python
#Left join to keep all the data, and we have to do it one at a time

merge_df=pd.merge(w_df, b_df, on='state', how='left')
merge_df=pd.merge(merge_df,i_df,on='state', how='left')
merge_df=pd.merge(merge_df,q_df, on='state', how='left')
merge_df.head()
```

```python
#Check columns
merge_df.columns
```

```python
#Check the shape
merge_df.shape
```

```python
#Collapse the city columns into one with the old 'fill then drop' technique.
merge_df['city']=merge_df['city_x'].fillna(merge_df['city_y'])
merge_df.drop(columns=['city_x', 'city_y'], inplace=True)
merge_df.head(10)
```

```python
#Check in on the shape
merge_df.shape
```

```python
#The old 'fill and drop' works for latitude and longitude as well

merge_df['latitude']=merge_df['latitude_x'].fillna(merge_df['latitude_y'])
merge_df.drop(columns=['latitude_x', 'latitude_y'], inplace=True)
merge_df['longitude']=merge_df['longitude_x'].fillna(merge_df['longitude_y'])
merge_df.drop(columns=['longitude_x', 'longitude_y'], inplace=True)
merge_df.head(10)
```

```python
#It's shaping up
merge_df.shape
```

```python
#Check the columns to see what can do
merge_df.columns
```

```python
#Get rid of some columns we don't need.
merge_df=merge_df.drop(columns=['Unnamed: 0_x', 'country_x', 'Unnamed: 0_y', 'Unnamed: 0', 'ID', 'country_y'])
merge_df.head(10)
```

```python
#Shaping up
merge_df.shape
```

```python
merge_df.columns
```

```python
#Check to see how many cities don't have a qol index
print(merge_df['qol_index'].isnull().sum())
```

```python
#Create the qol measure to determine what is 'high quality'.  totally subjective

qol_75=merge_df['qol_index'].quantile(0.69)

merge_df['high_q']=merge_df['qol_index'].apply(lambda x: 1 if x>=qol_75 else 0)

merge_df.head(10)
```

```python
#Check to see how many high quality cities there are
sum(merge_df['high_q'])
```

```python
# Drop duplicate cities
merge2_df=merge_df.drop_duplicates(subset=['city'])
merge2_df.shape
```

```python
#Now, how many high quality cities based on the subjective quartile
sum(merge2_df['high_q'])
```

```python
# Create the output file (CSV).
output_data_file = 'ml_data.csv'
# Export the data into a CSV.
merge2_df.to_csv(output_data_file)
```
