```python
import pandas as pd 
```

```python
cleanb_df=pd.read_csv('BEA_data_for_cross_reference.csv')
cleanb_df.head()
```

```python
cleanb_df=cleanb_df.rename(columns={'LineCode':'line_code','State': 'state', 'Description':'description'})
cleanb_df.head()
```

```python
cleanb_df=cleanb_df.drop(['line_code'], axis=1)
cleanb_df.head()
```

```python
# Create the output file (CSV).
output_data_file = 'BEA_data_for_cross_reference.csv'
# Export the data into a CSV.
cleanb_df.to_csv(output_data_file)
```

```python
cleanw_df=pd.read_csv('cities_weather_data_cleaned_us_only.csv')
cleanw_df.head()
```

```python
cleanw_df.columns
```

```python
cleanw_df=cleanw_df.drop(columns=['City_ID', 'City_ID.1'])
cleanw_df.head()
```

```python
cleanw_df=cleanw_df.rename(columns={'City':'city', 'State':'state', 'Country':'country', 'Date':'date', 'Lat':'latitude', 'Lng':'longitude', 'Max Temp':'max_temp', 'Humidity':'humidity', 'Cloudiness':'cloudiness', 'Wind Speed':'wind_speed'})
cleanw_df.head()
```

```python
# Create the output file (CSV).
output_data_file = 'cities_weather_data_cleaned_us_only.csv'
# Export the data into a CSV.
cleanw_df.to_csv(output_data_file)
```
