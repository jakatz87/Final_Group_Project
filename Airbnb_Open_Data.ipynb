```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read Airbnb_Open_Data.csv file
test_df = pd.read_csv('Airbnb_Open_Data.csv')
```

```python
test_df.head()
```

```python
test_df['total_price'] = test_df['total_price'].astype(int)
```

```python
print(test_df.columns)    
```

```python
test_df.count()
```

```python
# drop na's if 3 or more columns have blanks
test_df = test_df.dropna(thresh=3)
```

```python
# drop duplicates
test_df = test_df.drop_duplicates()
```

```python
test_df.count()
```

```python
# Calculate the average total price for each neighborhood
neighbourhood_avg_price = test_df.groupby('neighbourhood')['total_price'].mean()

# Create a bar chart of the average total price for each neighborhood
fig, ax = plt.subplots()
ax.bar(neighbourhood_avg_price.index, neighbourhood_avg_price.values)

# Set plot title and axis labels
ax.set_title("Average Total Price by neighbourhood")
ax.set_xlabel("neighbourhood")
ax.set_ylabel("Average Total Price")

# Rotate x-axis tick labels for better visibility
plt.xticks(rotation=90)

# Display the plot
plt.show()
```
