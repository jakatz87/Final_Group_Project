```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
#Read the file from GitHub raw
abb_cross_df=pd.read_csv('https://raw.githubusercontent.com/jakatz87/Final_Group_Project/main/ABnb_data_pull_2020_US.csv')
abb_cross_df.head(10)
```

```python
print(abb_cross_df['city'].nunique())
```

```python
abb_cross_df=abb_cross_df[['id','latitude', 'longitude', 'city', 'zipcode', 'bedrooms', 'bathrooms', 'accommodates', 'log_price', 'number_of_reviews','review_scores_rating']]
abb_cross_df.head(10)
```

```python
print(abb_cross_df['city'].nunique())
```

```python
#Use geopy to get the city and state
def get_state(lat, lon):
    geolocator = Nominatim(user_agent='my_app', timeout=20)
    location= geolocator.reverse(f"{lat}, {lon}")
    address= location.raw['address']
    return {'state': address.get('state', '')}

abb_cross_df[['state']]=abb_cross_df.apply(lambda row: pd.Series(get_state(row['latitude'], row['longitude'])), axis=1)
abb_cross_df.head(10)
```
