```python
import pandas as pd
```

```python
#read the file
qol_df=pd.read_csv('quality_of_life_data.csv')
qol_df.head()
```

```python
#Drop the empty top row
qol_df=qol_df.iloc[1:]
qol_df.head()
```

```python
#Split the city column into city, state, country
qol_df[['city','state','country']]=qol_df['city'].str.split(', ', expand=True)
qol_df.head()
```

```python
#Re-order the columns
qol_df=qol_df[['city_rank', 'city', 'state', 'country', 'qol_index', 'purchase_power', 'safety', 'healh_care', 'cost_of_living', 'prop_to_income_ratio', 'traffic_comm_time', 'pollution', 'climate']]
qol_df.head(20)
```

```python
#Keep only the US Cities
qol_df=qol_df[qol_df['country']=='United States']
qol_df.head(20)
```

```python
# Create the output file (CSV).
output_data_file = "qol_cleaned_us_only.csv"
# Export the data into a CSV.
qol_df.to_csv(output_data_file, index_label="ID")
```
