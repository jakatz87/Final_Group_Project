```python
import pandas as pd
```

```python
#Read the file from GitHub raw
inbound_cross_df=pd.read_csv('https://raw.githubusercontent.com/jakatz87/Final_Group_Project/main/Inbound%20Travellers%20State_data.csv')
inbound_cross_df.head(10)
```

```python
#Change the states to the abbreviations
state_dict = {'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA', 
              'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
              'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 
              'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD', 
              'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 
              'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 
              'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 
              'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 
              'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD', 'Tennessee': 'TN', 
              'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA', 
              'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY', 'District of Columbia': 'DC'}
inbound_cross_df['State']=inbound_cross_df['Dest State Nm'].map(state_dict)
inbound_cross_df=inbound_cross_df.drop(['Dest State Nm'], axis=1)
inbound_cross_df.head(10)
```

```python
#Drop any territories for consistency
inbound_cross_df=inbound_cross_df.dropna()
```

```python
#Reorder columns
inbound_cross_df=inbound_cross_df[['State','Latitude (generated)', 'Longitude (generated)', 'Sum PASSENGERS']]
inbound_cross_df.head(10)
```

```python
# Create the output file (CSV).
output_data_file = "inbound_data_for_cross_reference.csv"
# Export the data into a CSV.
inbound_cross_df.to_csv(output_data_file)
```
