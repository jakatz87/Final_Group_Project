```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
#Load the original file
clean_weather_df=pd.read_csv('cities_weather_data.csv')
clean_weather_df.head(10)
```

```python
#Keep only US cities
clean_weather_df=clean_weather_df[clean_weather_df['Country']=="US"]
clean_weather_df.head(10)
```

```python
#Use geopy to get states from lat and long
def get_state(lat,lon):
    geolocator = Nominatim(user_agent='my_app')
    location=geolocator.reverse(f"{lat}, {lon}")
    address=location.raw['address']
    state=address.get('state', '')
    return state
```

```python
#Create the 'State' column from the geopy code
clean_weather_df['State']=clean_weather_df.apply(lambda row: get_state(row['Lat'], row['Lng']), axis=1)
clean_weather_df.head(10)
```

```python
#Create a state dictionary to use abbreviations for cross referencing
state_dict = {'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA', 
              'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
              'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 
              'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD', 
              'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 
              'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 
              'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 
              'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 
              'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD', 'Tennessee': 'TN', 
              'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA', 
              'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'}

clean_weather_df['State']=clean_weather_df['State'].map(state_dict)
clean_weather_df.head(10)
```

```python
#Reorder columns
clean_weather_df=clean_weather_df[['City_ID', 'City', 'State', 'Country', 'Date', 'Lat', 'Lng', 'Max Temp', 'Humidity', 'Cloudiness', 'Wind Speed']]
clean_weather_df.head(10)
```

```python
# Create the output file (CSV).
output_data_file = "cities_weather_data_cleaned_us_only.csv"
# Export the data into a CSV.
clean_weather_df.to_csv(output_data_file, index_label="City_ID")
```
