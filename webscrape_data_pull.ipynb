```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import matplotlib.pyplot as plt
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the website
# https://www.numbeo.com/quality-of-life/region_rankings_current.jsp?region=021
url = 'https://www.numbeo.com/quality-of-life/region_rankings_current.jsp?region=021'
browser.visit(url)
```

```python
# Create a Beautiful Soup Object
html = browser.html
quality_soup = soup(html, 'html.parser')
```

```python
# Extract all rows of data
quality_data_rows = quality_soup.find_all('tr', role = 'row')
quality_data_rows
```

```python
# Create an empty list
data_list = []
# Loop through the scraped data to create a list of rows

for rows in quality_data_rows:
    row_data = rows.find_all('td')
    data_list.append(row_data)

data_list
    
```

```python
data2 = []
for row in data_list:
    tdlist = []
    for td in row:
        tdlist.append(td.text)
    data2.append(tdlist)
data2
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
row_heading = ['city_rank','city', 'qol_index', 'purchase_power', 'safety', 'healh_care', 'cost_of_living', 'prop_to_income_ratio', 'traffic_comm_time', 'pollution', 'climate']
quality_df = pd.DataFrame(data2, columns=row_heading)
quality_df.head()
```

```python
browser.quit()
```

```python
# Create the output file (CSV).
output_data_file = "quality_of_life_data.csv"
# Export the data into a CSV.
quality_df.to_csv(output_data_file, index_label="ID")
```
